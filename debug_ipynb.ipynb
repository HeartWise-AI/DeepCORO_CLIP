```python
import pandas as pd
import numpy as np
df_report  = pd.read_csv('data/reports/reports_with_alpha_separator_with_Calcifc_Stenosis_IFR_20250601_RCA_LCA_merged_with_left_dominance_dependent_vessels.csv', sep='α')
df_val_epoch = pd.read_csv('outputs/DeepCORO_clip/dev_deep_coro_clip_single_video/8av1xygm_20250605-083820_best_single_video/val_epoch0.csv')
merged_df_report = pd.merge(df_report, df_val_epoch, on='FileName', how='inner')

filtered_df = df_val_epoch[df_val_epoch['FileName'].str.contains('1.2.392.200036.9116.1467.20170720125102258.4.2', na=False)]
display(filtered_df)
# Define global variables
STUDY_INSTANCE_UID = filtered_df['StudyInstanceUID'].iloc[0]  # Get the first study instance UID
VESSEL = 'pda_stenosis'  # The vessel we're analyzing


# Get the study instance we're interested in
study_df = filtered_df.loc[(filtered_df.main_structure_name == 'Right Coronary')].reset_index(drop=True)

# Print initial debugging info
print("Study dataframe shape:", study_df.shape)
print("\nColumns available:", study_df.columns.tolist())

# Get the ground truth and predicted indices for all rows
gt_indices = study_df['ground_truth_idx']
pred_indices_list = [
    study_df['predicted_idx_1'],
    study_df['predicted_idx_2'],
    study_df['predicted_idx_3'],
    study_df['predicted_idx_4'],
    study_df['predicted_idx_5']
]

print("\nGround truth indices:", gt_indices.values)
print("Predicted indices shape:", len(pred_indices_list), "x", len(pred_indices_list[0]))

# Get the prox_rca_stenosis values for each index
gt_prox_rca_values = []
pred_prox_rca_values_list = []

for idx in range(len(study_df)):
    # Get ground truth value
    gt_prox_rca = merged_df_report.loc[merged_df_report['ground_truth_idx'] == gt_indices.iloc[idx], VESSEL].iloc[0]
    gt_prox_rca_values.append(gt_prox_rca)
    
    # Get predicted values for this row
    row_pred_values = []
    for pred_idx in [pred_indices_list[i].iloc[idx] for i in range(5)]:
        pred_prox_rca = merged_df_report.loc[merged_df_report['ground_truth_idx'] == pred_idx, VESSEL].iloc[0]
        row_pred_values.append(pred_prox_rca)
    pred_prox_rca_values_list.append(row_pred_values)

# Calculate means
row_means = [np.mean(row) for row in pred_prox_rca_values_list]
overall_mean = np.mean(row_means)
print("\nGround truth values for " + VESSEL + ":", gt_prox_rca_values)
print("Predicted " + VESSEL + " values per row:", pred_prox_rca_values_list)
print("\nMean predicted value per row:", row_means)
print("Overall mean predicted value:", overall_mean)

gt_stenosis = "gt_" + VESSEL
pred_stenosis = "pred_" + VESSEL
```
